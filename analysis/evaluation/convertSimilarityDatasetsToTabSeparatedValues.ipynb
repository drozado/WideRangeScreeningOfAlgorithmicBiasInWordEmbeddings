## Convert MEN similarity data set to Tab Separated form for usage with Gensim evaluate_word_pairs function

## Convert BATS evaluation data sets


```python
import itertools
import glob, os
#1_Inflectional_morphology, 2_Derivational_morphology, 3_Encyclopedic_semantics, 4_Lexicographic_semantics
inputPath = 'BATS_3.0/2_Derivational_morphology'
with open("2_Derivational_morphology.txt", "w") as fileOut:
    fileNames = [e for e in os.listdir(inputPath) if e.endswith('.txt')]
    for fileName in fileNames:
        print(fileName)
        fileNamePath=os.path.join(inputPath, fileName)
        with open(fileNamePath, "r") as fileIn:
            fileOut.write(': '+fileName+'\n')
            
            lines=[]
            for line in fileIn:
                lineModified = line.strip().replace('\t',' ')
                #print(lineModified)
                if '/' in line:
                    lineModified=lineModified.split('/', 1)[0]
                lines.append(lineModified)

            for element in itertools.permutations(lines,2):
                #print(element)
                e1 = element[0].split()[0]
                e2 = element[0].split()[1]
                e3 = element[1].split()[0]
                e4 = element[1].split()[1]
                outputLine =e1 + ' ' + e2 + ' ' + e3 + ' ' + e4 + '\n'
                fileOut.write(outputLine)
```
