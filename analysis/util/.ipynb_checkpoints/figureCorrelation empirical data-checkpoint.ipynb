## Plot real data on estimated axis from pole to pole

```python
def PlotDimensionH(ax,df, dim,constructPole1,constructPole2,fontSize=35,pole1Name=None,pole2Name=None,title="Axis"):
    ax.set_frame_on(False)
    ax.set_title(title, fontsize = 30,  y=10.08)
    colors = Coloring(df[dim])
    for i, word in enumerate(df.index):
        #r=random.randint(1,50)*0.1
        ax.annotate(word, (df[dim][i],0), color = colors[i], alpha = 0.6, fontsize = fontSize, rotation=60)
#         if i%2==0:
#             ax.annotate(word, (df[dim][i],0.5), color = colors[i], alpha = 0.6, fontsize = 25, rotation=40)
#         else:
#             ax.annotate(word, (df[dim][i],0), color = colors[i], alpha = 0.6, fontsize = 25, rotation=40)

    MaxY = df[dim].max()
    MinY = df[dim].min()
    
    
    bbox_props = dict(boxstyle="square", fc="blue", ec="blue",alpha=0.5, lw=0)
    if pole1Name==None:
        pole1Text = constructPole1[0]#'\n'.join(constructPole1)
    else:
        pole1Text = pole1Name
    #pole1Text = '\n'.join(constructPole1)
    ax.annotate(pole1Text, xy=(MinY,0),xytext=(MinY,2), color = "white", alpha = 1, fontsize = 20, rotation=0,bbox=bbox_props)
    bbox_props = dict(boxstyle="square", fc="red", ec="red", alpha=0.5,lw=0)
    if pole2Name==None:
        pole2Text = constructPole2[0]#'\n'.join(constructPole1)
    else:
        pole2Text = pole2Name
    ax.annotate(pole2Text, xy=(MaxY,1),xytext=(MaxY,2), color = "white", alpha = 1, fontsize = 20, rotation=0,bbox=bbox_props)    
    
    plt.xlim(MinY,MaxY)
    plt.axhline(0, color='black')
    plt.yticks(())
    plt.xticks(())
```

```python
fig = plt.figure(figsize = (150,0.1))
ax1 = fig.add_subplot(161)
PlotDimensionH(ax1, dataFrame, AxisName,constructPole1,constructPole2,
               fontSize=35,pole1Name="Masculine\nPole",pole2Name="Feminine\nPole",title="Gender Axis")
```

## Correlations

```python
%%R -o correlationPearson -o correlationSpearman -o correlationPvalue -o correlationTest 

# single correlation between a pair of variables
correlationPearson <- cor(Axis,RealValues,use="complete.obs",method="pearson")

#Spearman rho is a non-parametric statistic and so can be used when the data have violated parametric assumptions such
# as non-normally distributed data
correlationSpearman <- cor(Axis,RealValues,use="complete.obs",method="spearman")

# rcorr function for obtaining P values of correlations. Lives in packages Hmisc
correlationPvalue <- rcorr(Axis,RealValues,type="pearson")

# cor.test for obtaining confidence intervals and testing specific alternatives (two.sided, less, greater)
# alternative specifies whether you want to do a two-tailed test (alternative = two.
# sided), which is the default, or whether you predict that the correlation will be less
# than zero (i.e., negative) or more than zero (i.e., positive), in which case you can use
# alternative = less and alternative = greater, respectively.
correlationTest <- cor.test(Axis,RealValues, alternative = "two.sided", method = "pearson", conf.level = 0.95)
# correlationTest <- cor.test(Occupationsdf$GenderAxis,Occupationsdf$Percentages, alternative = "less", method = "pearson", conf.level = 0.95)
```

## t-SNE

```python
from sklearn.manifold import TSNE

from matplotlib.patches import Patch
from matplotlib.lines import Line2D

def visualize(px,rs,constructPole1,constructPole2):
    d= {}

#     constructPole1 = ['']
#     constructPole2 = ['']        
    
    cP1 =normalize(sum([normalize(model.wv[x]) for x in constructPole1]))
    cP2 =normalize(sum([normalize(model.wv[x]) for x in constructPole2]))
    
    model.wv['constructPole1']=cP1        
    model.wv['constructPole2']=cP2        

    constructPole1=['constructPole1']
    constructPole2=['constructPole2']
    
    RealDataTerms = list(RealData.keys())

    for t in constructPole1:
        d[t]=model.wv[t]

    for t in constructPole2:
        d[t]=model.wv[t]    

    for t in RealDataTerms:
        d[t]=model.wv[t]        
    
    vocab = list(d)
    vocabCategory = len(vocab)*['']
    for v in vocab:
        if v in constructPole1:
            i = vocab.index(v)
            vocabCategory[i]='constructPole1'
        if v in constructPole2:
            i = vocab.index(v)
            vocabCategory[i]='constructPole2'
        if v in RealDataTerms:
            i = vocab.index(v)
            vocabCategory[i]='RealData'
            
    #print("Number of data points: ", len(vocab))
    #print(" recall that perplexity should not be higher than this number")
    #print(vocabCategory)        

    X = model[vocab]

    tsne = TSNE(perplexity=px, n_components=2, learning_rate=5, init='pca',random_state=rs, n_iter=15000)

    X_tsne = tsne.fit_transform(X)

    df = pd.concat([pd.DataFrame(X_tsne),
                    pd.Series(vocab),pd.Series(vocabCategory)],
                   axis=1)

    df.columns = ['x', 'y', 'word','wordCategory']

    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(1, 1, 1)
    #filled_markers = ('o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X')
    for index, row in df.iterrows():
        #print(row["x"], row["y"],row["word"],row["wordCategory"])
        if row["wordCategory"]=='constructPole1':
            ax.scatter(row['x'], row['y'],c='r',s=100,marker='s')
            ax.annotate(row['word'], (row['x']+0.2, row['y']+0.2),size=25,color='r')
        if row["wordCategory"]=='constructPole2':
            ax.scatter(row['x'], row['y'],c='b',s=100,marker='o')
            ax.annotate(row['word'], (row['x']+0.05, row['y']),size=25,color='b')
        if row["wordCategory"]=='RealData':
            ax.scatter(row['x'], row['y'],c='k',s=100,marker='+', facecolors='none')        
            ax.annotate(row['word'], (row['x']+0.2, row['y']+0.2),size=25)


    legend_elements = [Line2D([0], [0], marker='s', color='w', label='constructPole1',markerfacecolor='r', markersize=15),
                   Line2D([0], [0], marker='o', color='w', markerfacecolor='b', label='constructPole2', markersize=15),
                   Line2D([0], [0], marker='+', color='k', label='RealData', markerfacecolor='k',markersize=15),
                    ]    
    plt.legend(handles=legend_elements,loc='upper left',prop={'size': 25})
        
    #ax.annotate(px, (0,0),size=15) #Reference
        
visualize(5,6,constructPole1,constructPole2)
```
