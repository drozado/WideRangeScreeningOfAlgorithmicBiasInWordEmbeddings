## Parsing lexicons

```python
# opinionLexicon
# https://www.cs.uic.edu/~liub/FBS/sentiment-analysis.html
import json
RealData = {}
with open('../listOfWords/opinion-lexicon-English/negative-words.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            w = l.strip().lower().replace('-','_')
#             print(w)
            RealData[w]=-1
            
with open('../listOfWords/opinion-lexicon-English/positive-words.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            w = l.strip().lower().replace('-','_')
#             print(w)
            RealData[w]=1            
print(json.dumps(RealData, indent = 4))
```

```python
#AFINN http://www2.imm.dtu.dk/pubdb/views/publication_details.php?id=6010
import json
RealData = {}
with open('../listOfWords/AFINN/AFINN-111.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            lineArray = l.strip().lower().split('\t')
            w = lineArray[0]
            sentiment = float(lineArray[1])
#             print(w,sentiment)
#             if w in model.wv:
            RealData[w]=sentiment
#                 print(w,sentiment)
print(json.dumps(RealData, indent = 4))
```

```python
#General inquirer http://www.wjh.harvard.edu/~inquirer/homecat.htm http://www.wjh.harvard.edu/~inquirer/spreadsheet_guide.htm
import json
RealData = {}
with open('../listOfWords/inquirer.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            lineArray = l.strip().lower().split()
            w = lineArray[0].replace('-','_')
            if '#' in w:
                w = w.split('#')[0]
            if w in RealData:
                continue
            else:
                preSent = lineArray[2]
    #             print(w,preSent)
                if preSent=="neg":
                    RealData[w]=-1
                elif preSent=="pos":
                    RealData[w]=1
                
print(json.dumps(RealData, indent = 4))
```

```python
#http://saifmohammad.com/WebPages/lexicons.html
import json
RealData = {}
with open('../listOfWords/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            lineArray = l.strip().lower().split('\t')
            w = lineArray[0]
#             print(w,lineArray[1],lineArray[2])
            if lineArray[1]=='negative' and lineArray[2] =='1':
                RealData[w]=-1
            elif lineArray[1]=='positive' and lineArray[2] =='1':                
                RealData[w]=1
print(json.dumps(RealData, indent = 4))
```

```python
#Obtaining list of positive and negative terms from Harvard inquirer
import pandas as pd
import numpy as np
import json
df = pd.read_csv('../listOfWords/inquirerbasic.csv')
RealData = {}
virtue = [w.lower().split('#')[0].replace('-','_') for w in list(df[['Entry','Virtue']].dropna()['Entry'])]
vice = [w.lower().split('#')[0].replace('-','_') for w in list(df[['Entry','Vice']].dropna()['Entry'])]

for w in vice:
    if w not in virtue: #Note that we are rejecting words with double mapping into vice and virtue
        RealData [w] = -1
for w in virtue:
    if w not in vice:
        RealData [w] = 1
print(json.dumps(RealData, indent = 4))    
```

```python
len(RealData)
```

```python
#Obtaining list of Affiliation and Hostile terms from Harvard inquirer
import pandas as pd
import numpy as np
import json
df = pd.read_csv('../listOfWords/inquirerbasic.csv')
RealData = {}
affiliation = [w.lower().split('#')[0].replace('-','_') for w in list(df[['Entry','Affil']].dropna()['Entry'])]
hostile = [w.lower().split('#')[0].replace('-','_') for w in list(df[['Entry','Hostile']].dropna()['Entry'])]

for w in hostile:
    if w not in affiliation: #Note that we are rejecting words with double mapping into Hostile and Affiliation
        RealData [w] = -1
for w in affiliation:
    if w not in hostile:
        RealData [w] = 1
print(json.dumps(RealData, indent = 4))    
```

## Read campaign contributions from different professions to Democrats and Republicans

```python
import json
 
# Open a json file into a json object
# Note that F.read() returns the content
# of the file as a string so we should call
# json.loads()
with open('../dataSources/politics_of_professions_data.json', encoding='utf-8') as F:
    data = json.loads(F.read())
 
```

```python
data
```

```python
professionPoliticalOrientation = {}
for professionalGroup in  data['children']:
    print(professionalGroup.keys())
    print("Professional group: ",professionalGroup['name'])
    for profession in professionalGroup['children']:
        professionName = profession['name'].lower().replace(' ','_')
        DRratio = profession['dems']/profession['reps']
        professionPoliticalOrientation[professionName]=DRratio
        print(professionName,DRratio)
```
