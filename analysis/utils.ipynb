## Creating list of Male and female names

```python
# opinionLexicon
# https://www.cs.uic.edu/~liub/FBS/sentiment-analysis.html
import json
RealData = {}
with open('./listOfWords/maleNames1960-1969.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            w = l.strip().replace('-','_').split(',')[0]
#             print(w)
            RealData[w]=-1
            
with open('./listOfWords/femaleNames1960-1969.txt') as f:
    for l in f.readlines():
        if len(l)>1:
            w = l.strip().replace('-','_').split(',')[0]
            if w in RealData: #for names in both lists ( names that are used for males and females)
                print(w)
                del(RealData[w])
            else:
                RealData[w]=1        

print(json.dumps(RealData, indent = 4))
```
